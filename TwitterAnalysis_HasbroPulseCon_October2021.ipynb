# Hasbro PulseCon 2021: Twitter Analysis

Hasbro PulseCon is a two-day virtual convention that toy company Hasbro, Inc. hosts to showcase exclusive products and reveal upcoming offerings for the next year to toy collectors. The virtual event took place on October 22 and 23, 2021. While the main consumers of their products are Children, part of their revenue comes from adult collectors. Featured brands include: Transformers, Star Wars, Power Rangers, NERF, Ghostbusters and GI Joe. 

This analysis measures the engagement and sentiment of tweets that featured the hashtag "#HasbroPulseCon". The tweets were downloaded on Sunday, October 24, 2021, in effort to get the hot take from fans. In terms of engagement, these tweets recieved 16.8 likes per tweet, 3.3 retweets, and 1.5 replies. 

```
+---------------------+---------------+
| ENGAGEMENT MEASURE  | VALUE         |
=======================================
| Likes               |16.8 per tweet |
+---------------------+---------------+
| Retweets            | 3.3 per tweet |
+---------------------+---------------+
| Replies             | 1.5 per tweet |
+---------------------+---------------+
```
There are two aspects of sentiment that were measured, Ploarity and Subjectivity. Polarity is the positive or negative score of the tweet on a scale of -1 (negative) to 1 (positive). The second metric, Subjectivity, is the fact or opinion score of the tweet. The Subjectivity scale is 0 (fact) to 1 (opinion). For collectors products like the ones showcased by Hasbro, the higher the subjectivity the better. This could be a good indication of how successfully the product is to making an emotional connection with the audience. This sample included 472 positive tweets compared to 91 negative tweets. The common perception of Twitter is that it is a very  negative platform, but the truth is that most tweets are positive in nature. People are often exposed to some level of negative responses and disagreement, therefore they tend to remember the negativity. The mean polarity score was 0.15 and the mean subjectivity score was 0.31. These are ok scores, but they are not particularly compelling. A polarity score closer to 0.2 and a subjectivity score over 0.4 would demonstrate a stronger interest. 
 
```
+--------------------+-------+---------+
| SENTIMENT MEASURE  | MEAN  | STD DEV |
========================================
| Polarity           | 0.148 | 0.275   |
+--------------------+-------+---------+
| Subjectivity       | 0.311 | 0.318   |
+--------------------+-------+---------+
```
Based on these results the consumer audience reaction for Hasbro PulseCon would be categorized as "Meets Expectations". The engagement metrics are extremely positive, and Hasbro is definately keeping their fan community engaged. 

````python
# Load Packages

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import json
````

````python
df = pd.read_json('HasbroPulseCon.json', lines = True)

pd.set_option('display.max_columns', None)

print("Data Shape:",df.shape)
df.head()
````

````python
# Create a new dataframes to subset the data by relavant attributes
tweetsRefined = pd.DataFrame(df['id'])

# Add the relevant columns
tweetsRefined.insert(1,"Date", df['date'])
tweetsRefined.insert(2,"Time", df['time'])
tweetsRefined.insert(3,"Language", df['language'])
tweetsRefined.insert(4,"Replies", df['replies_count'])
tweetsRefined.insert(5,"Likes", df['likes_count'])
tweetsRefined.insert(6,"Retweets", df['retweets_count'])
tweetsRefined.insert(7,"Tweet", df['tweet'])

# Preview the refined dataframe
tweetsRefined
````

````python
replies = sum(tweetsRefined['Replies'])
likes = sum(tweetsRefined['Likes'])
retweets = sum(tweetsRefined['Retweets'])

print('Replies:', replies)
print('Likes:', likes)
print('Retweets', retweets)
````

````python
# Import stopwords
import nltk
from nltk.corpus import stopwords

# Import textblob
from textblob import Word, TextBlob

nltk.download('stopwords')
nltk.download('wordnet')
stop_words = stopwords.words('english')
custom_stopwords = ['RT', '#HasbroPulseCon']
````

````python
# Create a function to clean each tweet
def preprocess_tweets(tweet, custom_stopwords):
    preprocessed_tweet = tweet
    preprocessed_tweet.replace('[^\w\s]','')
    preprocessed_tweet = " ".join(word for word in preprocessed_tweet.split() if word not in stop_words)
    preprocessed_tweet = " ".join(word for word in preprocessed_tweet.split() if word not in custom_stopwords)
    preprocessed_tweet = " ".join(Word(word).lemmatize() for word in preprocessed_tweet.split())
    return(preprocessed_tweet)

tweetsRefined['Processed Tweet'] = tweetsRefined['Tweet'].apply(lambda x: preprocess_tweets(x, custom_stopwords))
tweetsRefined.head()
````

````python
# Calculate sentiment
tweetsRefined['Polarity'] = tweetsRefined['Processed Tweet'].apply(lambda x: TextBlob(x).sentiment[0])
tweetsRefined['Subjectivity'] = tweetsRefined['Processed Tweet'].apply(lambda x: TextBlob(x).sentiment[1])

# Display descriptive statistics Sentiment Measurements
print('Polarity Statistics')
print(tweetsRefined['Polarity'].describe())
print('________________________________')
print('Subjectivity Statistic')
print(tweetsRefined['Subjectivity'].describe())
````

````python
# Create a dataframe for the 10 tweet moving average of polarity values
tweetMovAvg = tweetsRefined[['Date', 'Polarity']]
tweetMovAvg = tweetMovAvg.sort_values(by='Date', ascending=True)
tweetMovAvg['Moving Avg Polarity'] = tweetMovAvg.Polarity.rolling(10, min_periods=3).mean()

tweetMovAvg
````

````python
# Import Matplotlib
import matplotlib.pyplot as plt

# Create a visualization for the 10-tweet moving average
fig, ax = plt.subplots(figsize=(20, 5))
ax.plot(tweetMovAvg['Date'], tweetMovAvg['Moving Avg Polarity'], color='red')
plt.xticks(rotation = 90)


ax.set(xlabel='Date', ylabel='Moving Avg Polarity',
       title='Hasbro PulseCon: Moving Average Polarity')
ax.grid()

###### CHANGE SAVEFIG ######

fig.savefig("HasbroPulseCon.png")
plt.show()
````

````python
# Drill down by separating 'negative' tweets from 'positive' (Also eliminating neutral)
tweetsRefinedPOS = tweetsRefined[tweetsRefined['Polarity'] > 0]
tweetsRefinedNEG = tweetsRefined[tweetsRefined['Polarity'] < 0]

print("\nNumber of positive tweets: ", len(tweetsRefinedPOS))
print("\nNumber of negative tweets: ", len(tweetsRefinedNEG))
````

````python
# Preview the data sorted by Likes
tweetsEngagedPOS = tweetsRefinedPOS.sort_values('Likes', ascending=False)

tweetsEngagedPOS.head(5)
````
