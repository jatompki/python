# He-Man and The Master of the Universe: Twitter Analysis

Rumors of Mattel's plans to reboot the iconic He-Man and The Masters of the Universe (MOTU) toy land and media have been swirling for quite a while. In 2020, Mattel launched a retro toyline called Masters of the Universe Origins as a Walmart exclusive. As of January 2021, MOTU Origins became available at other retailers like Amazon and Target. For the most part the reception of these toys has been luke-warm. There are die-hard MOTU fans who have been eating this toyline up with a spoon, but the dozens of Skeletors hanging on the pegs is hard to ignore. The main audience for these toys, children, seem largely uninterested. In anticipation of two Netflix exclusive animated series, Mattel launched two new MOTU toy lines. Masterverse is a more modern collector's line that features the character designs of Netflix's Masters of the Universe Revelation. He-Man And The Masters Of The Universe Action Figures features re-designed versions of the characters as they appear in the Netflix series of the same name. These action figures are chunky, cartoony, and less detailed to appeal to children, four years-old and older (and a $10 to appeal to parents). It's too early to say how these more recent toy lines are doing, but needless to say that the popular reception of the Netflix series can give an indication about how much buzz there is about this property. The analysis below focuses on the redesigned "For Children" Netflix sereies He-Man And The Masters Of The Universe which was released on Thursday, September 16, 2021.

This analysis measures the engagement and sentiment of tweets that featured the hashtag "#HeManandtheMastersoftheUniverse". The tweets were downloaded on Monday, September 20, 2021, in effort to get the hot take from fans and audience members. In terms of engagement, these tweets recieved 6.6 likes per tweet, 1.0 retweets, and 0.6 replies. 

```

+---------------------+---------------+

+---------------------+---------------+
```
There are two aspects of sentiment that were measured Polarity, the positive or negative score of the tweet. The Polarity scale is -1 (negative) to 1 (positive). The second metric is Subjectivity, the fact or opinion score of the tweet. The Subjectivity scale is 0 (fact) to 1 (opinion). For a media property like Masters of the Universe, the higher the subjectivity the better. This could be a good indication of how successfully the property is to making an emotional connection with the audience. The MOTU series had 210 positive tweets compared to 27 negative tweets. The common perception of Twitter is that it is such a negative platform, but the truth is that most tweets are positive in nature. People of often exposed to some level of negative responses and disagreement, therefore they tend to remember the negativity. The mean polarity score was 0.15 and the mean subjectivity score was 0.33. Neither of these scores are too bad, but they are also not very compelling. I would like to see Polarity climb closer to 0.2 and Subjectivity increase to over 0.4. Based on these results, I would determine that the mass audience reception for MOTU is still luke warm. I would expect MOTU to struggle as a brand until there is a much stronger media presence behind it.

````python
# Load Packages

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import json
````

````python
df = pd.read_json('HeMan_Sept2021.json', lines = True)

pd.set_option('display.max_columns', None)

print("Data Shape:",df.shape)
df.head()
````

````python
# Create a new dataframes to subset the data by relavant attributes
tweetsRefined = pd.DataFrame(df['id'])

# Add the relevant columns
tweetsRefined.insert(1,"Date", df['date'])
tweetsRefined.insert(2,"Time", df['time'])
tweetsRefined.insert(3,"Language", df['language'])
tweetsRefined.insert(4,"Replies", df['replies_count'])
tweetsRefined.insert(5,"Likes", df['likes_count'])
tweetsRefined.insert(6,"Retweets", df['retweets_count'])
tweetsRefined.insert(7,"Tweet", df['tweet'])

# Preview the refined dataframe
tweetsRefined
````

````python
replies = sum(tweetsRefined['Replies'])
likes = sum(tweetsRefined['Likes'])
retweets = sum(tweetsRefined['Retweets'])

print('Replies:', replies)
print('Likes:', likes)
print('Retweets', retweets)
````

````python
# Import stopwords
import nltk
from nltk.corpus import stopwords

# Import textblob
from textblob import Word, TextBlob

nltk.download('stopwords')
nltk.download('wordnet')
stop_words = stopwords.words('english')
custom_stopwords = ['RT', '#heman', '#HeMan', '#motu', '#MastersoftheUniverse', '#HeManandtheMastersoftheUnivers']
````

````python
# Create a function to clean each tweet
def preprocess_tweets(tweet, custom_stopwords):
    preprocessed_tweet = tweet
    preprocessed_tweet.replace('[^\w\s]','')
    preprocessed_tweet = " ".join(word for word in preprocessed_tweet.split() if word not in stop_words)
    preprocessed_tweet = " ".join(word for word in preprocessed_tweet.split() if word not in custom_stopwords)
    preprocessed_tweet = " ".join(Word(word).lemmatize() for word in preprocessed_tweet.split())
    return(preprocessed_tweet)

tweetsRefined['Processed Tweet'] = tweetsRefined['Tweet'].apply(lambda x: preprocess_tweets(x, custom_stopwords))
tweetsRefined.head()
````

````python
# Calculate sentiment
tweetsRefined['Polarity'] = tweetsRefined['Processed Tweet'].apply(lambda x: TextBlob(x).sentiment[0])
tweetsRefined['Subjectivity'] = tweetsRefined['Processed Tweet'].apply(lambda x: TextBlob(x).sentiment[1])

# Display descriptive statistics Sentiment Measurements
print('Polarity Statistics')
print(tweetsRefined['Polarity'].describe())
print('________________________________')
print('Subjectivity Statistic')
print(tweetsRefined['Subjectivity'].describe())
````

````python
# Create a dataframe for the 10 tweet moving average of polarity values
tweetMovAvg = tweetsRefined[['Date', 'Polarity']]
tweetMovAvg = tweetMovAvg.sort_values(by='Date', ascending=True)
tweetMovAvg['Moving Avg Polarity'] = tweetMovAvg.Polarity.rolling(10, min_periods=3).mean()

tweetMovAvg
````

````python
# Import Matplotlib
import matplotlib.pyplot as plt

# Create a visualization for the 10-tweet moving average
fig, ax = plt.subplots(figsize=(20, 5))
ax.plot(tweetMovAvg['Date'], tweetMovAvg['Moving Avg Polarity'], color='purple')
plt.xticks(rotation = 90)


ax.set(xlabel='Date', ylabel='Moving Avg Polarity',
       title='Masters of the Universe: Moving Average Polarity')
ax.grid()

fig.savefig("heman_Sept2021.png")
plt.show()
````

````python
# Drill down by separating 'negative' tweets from 'positive' (Also eliminating neutral)
tweetsRefinedPOS = tweetsRefined[tweetsRefined['Polarity'] > 0]
tweetsRefinedNEG = tweetsRefined[tweetsRefined['Polarity'] < 0]

print("\nNumber of positive tweets: ", len(tweetsRefinedPOS))
print("\nNumber of negative tweets: ", len(tweetsRefinedNEG))
````

````python
# Preview the data sorted by Likes
tweetsEngagedPOS = tweetsRefinedPOS.sort_values('Likes', ascending=False)

tweetsEngagedPOS.head(5)
````
